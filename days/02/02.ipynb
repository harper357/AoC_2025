```python
with open('example.txt') as f:
    example_input = f.read().splitlines()[0]
```

Initial idea:
split range numbers, is even? split the number, check if the same?


```python
def split_input(input_string: str) -> list[str]:
    split_string = input_string.split(",")
    input_array = [x.split("-") for x in split_string]
    return pd.DataFrame(input_array, columns=["start", "stop"], dtype=str)
```

```python
def find_first_halfs(start: str, stop: str):
    start_len = len(start)
    stop_len = len(stop)
    if start_len % 2 == 0:
        start_half = start[: int(start_len / 2)]
    elif start_len == 1:
        start_half = 0
    else:
        start_half = start[: math.floor(start_len / 2)]
    if stop_len % 2 == 0:
        stop_half = stop[: int(stop_len / 2)]
    else:
        stop_half = stop[: math.ceil(stop_len / 2)]
    return list(range(int(start_half), int(stop_half) + 1))
```

```python
def find_min_halfhalfs(start: str, canidates_list: list):
    start_int = int(start)
    for n, each in enumerate(canidates_list):
        if int(str(each) + str(each)) >= start_int:
            return n
    return len(canidates_list)


def find_max_halfhalfs(stop: str, min_n, canidates_list: list):
    stop_int = int(stop)
    for n, each in enumerate(canidates_list[min_n:], min_n):
        if int(str(each) + str(each)) > stop_int:
            return n
    return len(canidates_list)


def find_valids_sum(start, stop, canidates_list):
    min_n = find_min_halfhalfs(start, canidates_list)
    max_n = find_max_halfhalfs(stop, min_n, canidates_list)
    valids = [int(str(x) + str(x)) for x in canidates_list[min_n:max_n]]
    return sum(valids)
```

```python
input_df = split_input(example_input)
display(input_df, input_df.dtypes)
```

```python
for index, each in input_df.iterrows():
    (find_first_halfs(each["start"], each["stop"]))
```

```python
input_df["sums"] = input_df.apply(
    lambda x: find_valids_sum(
        x["start"], x["stop"], find_first_halfs(x["start"], x["stop"])
    ), axis=1
)
```

## part 2

```python
def check_number(number):
    number_str = str(number)
    for n in range(1,len(number_str)//2+1):
        if len(set(itertools.batched(number_str, n))) == 1:
            return number
    return 0
```

```python
# example inputs
input2_df = split_input(example_input)
input2_df["sums"] = input2_df.apply(
    lambda x: sum([check_number(y) for y in range(int(x["start"]), int(x["stop"])+1)]),
    axis=1,
)
input2_df["sums"].sum()
```
