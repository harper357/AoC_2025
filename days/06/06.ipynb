```python
def read_d06_v2(filepath):
    with open(filepath, "r") as f:
        problems = f.read().splitlines()
    return problems
```

```python
def format_and_eval_v2(input):
    input = np.array([list(x[::-1]) for x in input]).T
    input = ["".join(x) for x in input]
    formatted = []
    mini = []
    for x in input:
        mini.append(x)
        if x.endswith('*') or x.endswith('+'):
            formatted.append(mini)
            mini = []
    formatted = [''.join(x) for x in formatted]
    formatted = [x[-1].join(x[:-1].split()) for x in formatted]
    sums = sum([eval(x) for x in formatted])
    return sums
```

```python
example2 = np.array(read_d06_v2("example.txt"))
```

### input

```python
evalled_input = format_and_eval_v2(example2)
assert(evalled_input==3263827)
```

```python
input2 = np.array(read_d06_v2("input.txt"))
assert(format_and_eval_v2(input2)==11708563470209)
```
