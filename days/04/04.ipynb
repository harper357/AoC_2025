## Example

```python
with open('example.txt') as f:
    example_input = f.read().splitlines()
with open('example_output.txt') as f:
    example_output = f.read().splitlines()
```

```python
def check_around(row, col, df):
    around_mods = [-1, 0, 1]
    shape = df.shape
    count = 0
    for r, c in it.product(around_mods, around_mods):
        new_r = row + r
        new_c = col + c
        if (0 <= new_r < shape[0]) & (0 <= new_c < shape[1]):
            if r == c == 0:
                next
            else:
                if df.iat[new_r, new_c] == "@":
                    count += 1
    return count


def check_elements(df, count_limit):
    shape = df.shape
    new_array = np.empty_like(df.values)
    idx_gr = it.product(range(shape[0]), range(shape[1]))
    for row, col in idx_gr:
        if df.iat[row, col] == "@":
            count = check_around(row, col, df)
            if count < count_limit:
                new_array[row, col] = 1
    return new_array
```

```python
example_df = pd.DataFrame([list(x) for x in example_input])
example_test_df = pd.DataFrame(check_elements(example_df, 4))
```

```python
with open("input.txt") as f:
    inputs = [list(x.strip()) for x in f.readlines()]
input_df = pd.DataFrame(inputs)
output_df = pd.DataFrame(check_elements(input_df, 4))
output_df.sum().sum()
```

## part 2


```python
def sum_removed_and_update_df(df, removal_df):
    sum = removal_df.sum().sum()
    if sum > 0:
        new_df = df.where(~removal_df.astype(bool), ".")
        return new_df, sum
    else:
        return None, sum
```

### example

```python
removed_counts = []
example_2_df = example_df
while example_2_df is not None:
    removal_df = pd.DataFrame(check_elements(example_2_df, 4))
    example_2_df, sum_count = sum_removed_and_update_df(example_2_df, removal_df)
    # print(sum_count)
    removed_counts.append(sum_count)
print(sum(removed_counts))
```

### inputs

```python
removed_counts = []
output_2_df = input_df
while output_2_df is not None:
    removal_df = pd.DataFrame(check_elements(output_2_df, 4))
    output_2_df, sum_count = sum_removed_and_update_df(output_2_df, removal_df)
    removed_counts.append(sum_count)
print(sum(removed_counts))
```
