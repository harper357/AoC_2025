```python
import pandas as pd
```

### example


```python
with open('example.txt') as f:
    example_input = f.read().splitlines()
```

```python
example_df = pd.DataFrame(
    [list(x) for x in example_input],
)
```

```python
def find_top_two(series):
    idxmax = series[:-1].idxmax()
    second_idxmax = series[idxmax + 1 :].idxmax()
    top_two = ("").join(series[[idxmax, second_idxmax]].values)
    return top_two
```

```python
example_df["top_two"] = example_df.apply(find_top_two, axis="columns")
```

```python
example_df["top_two"].astype(int).sum()
```

## part 1


```python
with open("input.txt") as f:
    input = f.readlines()
    input = [list(x.strip()) for x in input]
input_df = pd.DataFrame(input)
input_df.shape
```

```python
input_df["top_two"] = input_df.apply(find_top_two, axis="columns")
input_df["top_two"].astype(int).sum()
```

## part 2


```python
def find_top_n(series, n):
    idx_list = [series[0 : -(n - 1)].idxmax()]
    for idx in range(1, n):
        idx_start = idx_list[-1] + 1
        idx_stop = -(n - idx) + 1
        if idx_stop == -0:
            short_series = series[idx_start:]
        else:
            short_series = series[idx_start:idx_stop]
        new_idx_max = short_series.idxmax()
        # display([idx_start, idx_stop, short_series.to_list()])
        idx_list.append(new_idx_max)
    top_n = ("").join(series[idx_list].values)
    return top_n
```

### example

```python
find_top_n(pd.Series(list(example_input[0])), n=12)
```

```python
example_df["part_2_sum"] = example_df.drop(
    ["top_two"], axis="columns"
).apply(find_top_n, n=12, axis="columns")
```

```python
example_df["part_2_sum"].astype(int).sum()
```

### inputs


```python
input_df["part_2_sum"] = input_df.drop(["top_two"], axis="columns").apply(
    find_top_n, n=12, axis="columns"
)
input_df["part_2_sum"].astype(int).sum()
```
